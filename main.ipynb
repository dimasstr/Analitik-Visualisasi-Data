```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap
```

```python
#lihat Data
namaFile = 'kuisioner.csv'
data = pd.read_csv(namaFile)
data
```

```python
#Menghapus secara Horizontal
data_any = data.dropna(axis=0, how='any')
data_any.shape[0]
data_any
```

```python
#Menghapus data yang duplicate
dplcd = data_any.append(data_any.iloc[[-2]])
dplcd
```

```python
x = dplcd.drop_duplicates()
x
```

```python
pieChart = x['Apakah Anda pernah terpapar hepatitis?'].value_counts(normalize=True)

fig,bar = plt.subplots(figsize = (5,9))
bar = pieChart.plot.pie(figsize =(10,20), autopct = "%.2f%%")
bar.set_title('Persentase Responden yang pernah terpapar hepatitis A',fontdict={'fontsize': 18, 'fontweight':'bold'}, y = 1)
bar.set_xlabel('Persentase', fontdict={'fontsize': 18, 'fontweight':'bold'})
bar.set_ylabel(' ', fontdict={'fontsize': 18, 'fontweight':'bold'})
```

```python
# Sertifikasi sesuai kualifikasinya
fig, bar = plt.subplots(figsize = (11,7))
bar = sns.countplot(x='Asal Kecamatan',data=dplcd)
bar.set_title('DATA ASAL RESPONDEN',fontdict= { 'fontsize': 24, 'fontweight':'bold'}, y = 1.05)
bar.set_xlabel('Daerah',fontdict= { 'fontsize': 20, 'fontweight':'bold'} )
bar.set_ylabel('Banyak Data',fontdict= { 'fontsize': 20, 'fontweight':'bold'} )
```
